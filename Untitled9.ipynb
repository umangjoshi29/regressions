```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
tips = sns.load_dataset("tips")
```

```python
#Display the first few rows to understand database
print(tips.head())
```

```python
#Selecting feature for simple linear regression 
#We will predict "tip"(dependent variable) based on total
x= tips[["total_bill"]] #(independent variable)
y= tips[["tip"]] # dependent variable
```

```python
# split the dataset into training and testing sets(80-20)
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#Fit simple Linear Regression Model
model= LinearRegression()
model.fit(x_train,y_train)
```

```python
m = model.coef_[0]  # slope (assuming it's a single-element array)
b = model.intercept_  # intercept
print(f"Regression Equation: Y = {b} + {m}X")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f"Mean Squared Error(MSE):{mse:.2f}")
print(f"R-squared (R2):{r2:2f}")
```

```python
plt.figure(figsize = (8,5))
plt.scatter(x_test, y_test, color = 'blue', label = "Actual Tips")
plt.plot(x_test, y_pred, color = 'red', label = "Regression Line")
plt.xlabel = ('Total Bill')
plt.ylabel = ('Tip')
plt.title('Simple Linear Regression: Total Bill vs Tip')
plt.legend()
plt.show()
```

```python
#multiple regression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd

# Load dataset
tips = sns.load_dataset("tips")

# Convert 'sex' to numeric values (0: Female, 1: Male) for visualization
tips['sex_numeric'] = tips['sex'].apply(lambda x: 1 if x == 'Male' else 0)

# Create 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with meaningful axes
ax.scatter(tips['total_bill'], tips['tip'], tips['sex_numeric'], c=tips['sex_numeric'], cmap='coolwarm', alpha=0.7)

# Set axis labels
ax.set_xlabel("Total Bill ($)")  
ax.set_ylabel("Tip ($)")         
ax.set_zlabel("Sex (0 = Female, 1 = Male)")  

plt.title("3D Multiple Regression Visualization")
plt.show()
```
